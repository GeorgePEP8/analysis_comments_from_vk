# Импорт данных

```python
import joblib
import pandas as pd
import re
import pickle

# загрузка модели
filename = 'for_use_model/finalized_model.pkl'
count_idf = joblib.load('count_idf.pkl')
loaded_model = joblib.load(filename)
```

```python
with open("comments/comments.txt", "r", encoding='utf-8') as file:      # file = open("myfile.txt")
        comments = file.read().split('\n')
del comments[-1]
len(comments)
```

```python
with open("comments/comments_clear.txt", "r", encoding='utf-8') as file:      # file = open("myfile.txt")
        comments_clear = file.read().split(',')
del comments_clear[-1]
comments_clear = [i.replace("\n", "") for i in comments_clear]
comments_clear
```

```python
data_set = pd.DataFrame({'comments':comments, 'comments_clear': comments_clear})
data_set
```

# Вывод оценки рандомных комментов

```python
for _ in range(5):

    source = data_set.sample(n=1)
    text_clear = source['comments_clear'].values[0]
    text = source['comments'].values[0]

    print(text)

    tf_idf_text = count_idf.fit_transform([text_clear])

    toxic_proba = loaded_model.predict_proba(tf_idf_text)

    print('Вероятность негатива: ', toxic_proba[:, 0])
    print()
```

# Вывод оценки выборочных комментов

```python

for i in range(6):

    source = data_set.iloc[i]
    text_clear = source['comments_clear']
    text = source['comments']

    print(text)

    tf_idf_text = count_idf.fit_transform([text_clear])

    toxic_proba = loaded_model.predict_proba(tf_idf_text)

    print('Вероятность негатива: ', toxic_proba[:, 0])
    print()
```

```python

count_split = []
comment_spilt = []
tonality = []
```

```python
# функция подсчета кол-во разбиений. return кол-во разбиений, сами разбиения
def sent_count(text: str) -> (int, list):
    new_text = re.sub(r'[,.!?]\s', r'|', text) 
    sent_num = len(new_text.split('|') )   
    print('В этом тексте {} разбиений.'.format(sent_num))
    return sent_num, new_text.split('|')

for comment in comments:
    a,b = sent_count(comment)
    count_split.append(a)
    comment_spilt.append(b)
```

```python
tonality = []
for i in range(0, len(comments)):

    source = data_set.iloc[i]
    text_clear = source['comments']
    text = source['comments']

    print(text)

    tf_idf_text = count_idf.fit_transform([text_clear])

    toxic_proba = loaded_model.predict_proba(tf_idf_text)

    print('Вероятность негатива: ', toxic_proba[:, 0])
    print()
    tonality.append(toxic_proba[:, 0])
tonality
```
